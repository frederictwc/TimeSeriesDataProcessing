```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data/bitcoin_historical.csv')
data = data.dropna()
```

```python
plt.plot(data['Timestamp'],data['Close'],linewidth=0.5)
plt.title('USD/Bitcoin Historical')
plt.xlabel('Time')
plt.ylabel('USD/Bitcoin')
plt.show()
```
